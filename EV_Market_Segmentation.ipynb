```python
# Importing libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import re
import numpy as np
import matplotlib.pyplot as plt
# Load the preprocessed data from the CSV file
df_electriccardata = pd.read_csv('/content/drive/MyDrive/Prev/preprocessed_electriccardata.csv')

# Function to extract numerical values from strings using regular expressions
# Function to extract numerical values from strings using regular expressions
def extract_numeric_value(value):
    try:
        # Extract numeric values with optional decimal point and unit (km, km/h, sec)
        numeric_value = float(re.findall(r'(\d+(\.\d+)?) (km/h|km|sec)', str(value))[0][0])
    except (ValueError, IndexError):
        numeric_value = np.nan
    return numeric_value
# Apply the function to extract numeric values from 'Accel', 'TopSpeed', and 'Range' columns
df_electriccardata['Accel'] = df_electriccardata['Accel'].apply(extract_numeric_value)
df_electriccardata['TopSpeed'] = df_electriccardata['TopSpeed'].apply(extract_numeric_value)
df_electriccardata['Range'] = df_electriccardata['Range'].apply(extract_numeric_value)

# Drop rows with NaN values
df_electriccardata.dropna(subset=['Accel', 'TopSpeed', 'Range'], inplace=True)
# Check if the DataFrame contains any rows
if df_electriccardata.shape[0] == 0:
    raise ValueError("No valid data remains after handling non-numeric values.")
# Feature Engineering: Scale the data for clustering
scaler = StandardScaler()
scaled_data_electriccardata = scaler.fit_transform(df_electriccardata[['Accel', 'TopSpeed', 'Range']])

# Choosing the Number of Clusters: Elbow Method
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data_electriccardata)
    inertia.append(kmeans.inertia_)

# Plotting Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-cluster Sum of Squares (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()

# Applying K-means clustering with the optimal number of clusters
k = 3  # Optimal number of clusters based on Elbow Method
kmeans = KMeans(n_clusters=k, random_state=42)
df_electriccardata['Segment'] = kmeans.fit_predict(scaled_data_electriccardata)

# Visualization: Plotting clusters based on two dimensions (PCA)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data_electriccardata = pca.fit_transform(scaled_data_electriccardata)

plt.figure(figsize=(8, 6))
plt.scatter(pca_data_electriccardata[:, 0], pca_data_electriccardata[:, 1], c=df_electriccardata['Segment'], cmap='rainbow')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering (electriccardata_norm.csv)')
plt.grid(True)
plt.show()

# Segment Validation: Analyzing cluster characteristics
cluster_characteristics_electriccardata = df_electriccardata.groupby('Segment').mean()

# Segment Naming: Based on cluster characteristics and attributes
segment_names_electriccardata = {0: 'Economic', 1: 'Performance', 2: 'Versatile'}

# Final Results: Assigning segment names to each data point
df_electriccardata['Segment_Name'] = df_electriccardata['Segment'].map(segment_names_electriccardata)

# Save the final segmented data to a new CSV file
df_electriccardata.to_csv('/content/drive/MyDrive/Prev/segmented_electriccardata.csv', index=False)
```

```python
# Importing libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the preprocessed data from the CSV file
df_EVstats = pd.read_csv('/content/drive/MyDrive/Prev/preprocessed_EVstats.csv')

# Handling Missing Values: Filling NaN values with the mean
df_EVstats.fillna(df_EVstats.mean(), inplace=True)

# Feature Engineering: Scale the data for clustering
scaler_EVstats = StandardScaler()
scaled_data_EVstats = scaler_EVstats.fit_transform(df_EVstats[['Passenger Cars (Category M1 as per CMVR)', 'Buses']])

# Choosing the Number of Clusters: Elbow Method
inertia_EVstats = []
k_range_EVstats = range(1, 11)
for k in k_range_EVstats:
    kmeans_EVstats = KMeans(n_clusters=k, random_state=42)
    kmeans_EVstats.fit(scaled_data_EVstats)
    inertia_EVstats.append(kmeans_EVstats.inertia_)

# Plotting Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_range_EVstats, inertia_EVstats, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-cluster Sum of Squares (Inertia)')
plt.title('Elbow Method for Optimal k (EVstats.csv)')
plt.grid(True)
plt.show()

# Applying K-means clustering with the optimal number of clusters
k_EVstats = 3  # Optimal number of clusters based on Elbow Method
kmeans_EVstats = KMeans(n_clusters=k_EVstats, random_state=42)
df_EVstats['Segment'] = kmeans_EVstats.fit_predict(scaled_data_EVstats)

# Segment Naming
segment_names_EVstats = {0: 'Segment 0', 1: 'Segment 1', 2: 'Segment 2'}
df_EVstats['Segment_Name'] = df_EVstats['Segment'].map(segment_names_EVstats)

# Save the final segmented data to a new CSV file
df_EVstats.to_csv('/content/drive/MyDrive/Prev/segmented_EVstats.csv', index=False)
```

```python
# Importing libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the preprocessed data from the CSV file
df_charging_station = pd.read_csv('/content/drive/MyDrive/Prev/preprocessed_charging_station.csv')

# Feature Engineering: Scale the data for clustering
df_charging_station['power'] = df_charging_station['power'].apply(lambda x: re.findall(r'\d+', x)[0]).astype(float)
scaler_charging_station = StandardScaler()
scaled_data_charging_station = scaler_charging_station.fit_transform(df_charging_station[['latitude', 'longitude', 'power']])

# Choosing the Number of Clusters: Elbow Method
inertia_charging_station = []
k_range_charging_station = range(1, 11)
for k in k_range_charging_station:
    kmeans_charging_station = KMeans(n_clusters=k, random_state=42)
    kmeans_charging_station.fit(scaled_data_charging_station)
    inertia_charging_station.append(kmeans_charging_station.inertia_)

# Plotting Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_range_charging_station, inertia_charging_station, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-cluster Sum of Squares (Inertia)')
plt.title('Elbow Method for Optimal k (electric_vehicle_charging_station.csv)')
plt.grid(True)
plt.show()

# Applying K-means clustering with the optimal number of clusters
k_charging_station = 3  # Optimal number of clusters based on Elbow Method
kmeans_charging_station = KMeans(n_clusters=k_charging_station, random_state=42)
df_charging_station['Segment'] = kmeans_charging_station.fit_predict(scaled_data_charging_station)

# Segment Naming
segment_names_charging_station = {0: 'Segment 0', 1: 'Segment 1', 2: 'Segment 2'}
df_charging_station['Segment_Name'] = df_charging_station['Segment'].map(segment_names_charging_station)

# Save the final segmented data to a new CSV file
df_charging_station.to_csv('/content/drive/MyDrive/Prev/segmented_charging_station.csv', index=False)
```
